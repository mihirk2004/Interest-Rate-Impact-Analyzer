```python
import pandas as pd

# Load the CSV file
csv_file = "loan.csv"  # Replace with your file path
df = pd.read_csv(csv_file, header=None)  # Read without headers

# Select two specific rows (e.g., row index 0 and 1) and transpose them
row1 = df.iloc[0].reset_index(drop=True)  # First row
row2 = df.iloc[1].reset_index(drop=True)  # Second row

# Combine into a new DataFrame with two columns
df_transposed = pd.DataFrame({'Column1': row1, 'Column2': row2})

# Save as a new CSV file
df_transposed.to_csv("output_transposed.csv", index=False, header=False)

print("Transposed CSV file saved as output_transposed.csv")
```

```python
import pandas as pd

# Load the CSV file
csv_file = "loan.csv"  # Replace with your actual file name
df = pd.read_csv(csv_file)

# Specify the column to modify
column_name = "Date"  # Replace with the actual column name

# Remove the last 5 characters from each row in the specified column
df[column_name] = df[column_name].astype(str).str[:-9]
# df = df.round(3)
# Save the updated CSV file
df.to_csv("updated_file.csv", index=False)

print("Updated CSV file saved as updated_file.csv")
```
